```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded successfully")
```

```python
df = pd.read_csv("C:/data-analysis-portfolio/datasets/weather data.csv")
print("Data loaded successfully")
```

```python
print("Dataset Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
df = df.fillna(df.mean(numeric_only=True))

print("Missing values handled")
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df = df.rename(columns={
'datetime_utc':'date',
'_tempm':'temperature',
'_hum':'humidity',
'_pressurem':'pressure',
'_wspdm':'wind_speed'
})
```

```python
print("\nStatistical Summary")
print(df.describe())
```

```python
plt.figure(figsize=(10,5))

plt.plot(df["date"], df["temperature"])

plt.title("Temperature Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.savefig("C:/data-analysis-portfolio/visualizations/Temperature_Trend.png")


plt.show()
```

```python
df["month"] = df["date"].dt.month

monthly_temp = df.groupby("month")["temperature"].mean()

print(monthly_temp)
```

```python
df["month"] = df["date"].dt.month

monthly_temp = df.groupby("month")["temperature"].mean()

print(monthly_temp)
```

```python
df["month"] = df["date"].dt.month

monthly_temp = df.groupby("month")["temperature"].mean()

print(monthly_temp)
```

```python
plt.figure(figsize=(8,5))

monthly_temp.plot(kind="bar")

plt.title("Average Temperature by Month")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.savefig("C:/data-analysis-portfolio/visualizations/Average Temperature by Month.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["wind_speed"])

plt.title("Wind Speed Distribution")
plt.savefig("C:/data-analysis-portfolio/visualizations/Wind Speed Distribution.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["pressure"], bins=20)

plt.title("Pressure Distribution")
plt.savefig("C:/data-analysis-portfolio/visualizations/Pressure Distribution.png")
plt.show()
```

```python
sns.histplot(df['humidity'], bins=30)
plt.title("Humidity Distribution")
plt.savefig("C:/data-analysis-portfolio/visualizations/Humidity Distribution.png")
plt.show()
```

```python
correlation = df[['temperature','humidity','wind_speed','pressure']].corr()

print(correlation)
```

```python
corr = df[['temperature','humidity','pressure','wind_speed']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Weather Correlation")
plt.savefig("C:/data-analysis-portfolio/visualizations/Weather Correlation.png")
plt.show()
```
