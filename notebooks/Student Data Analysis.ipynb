```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("C:/data-analysis-portfolio/datasets/student_data.csv")
print(df.head())
```

```python
print("shape:",df.shape)
print("\nColumns:")
print(df.columns)
print("\nMissing values:")
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
avg_scores = df[['G1','G2','G3']].mean()

print(avg_scores)
```

```python
avg_scores.plot(kind='bar')

plt.title("Average Student Grades")
plt.ylabel("Average Score")
plt.savefig("C:/data-analysis-portfolio/visualizations/Average Student Grades.png")
             

plt.show()
```

```python
df["Pass"] = df["G3"] >= 10

pass_count = df["Pass"].value_counts()

print(pass_count)
```

```python
pass_count.plot(kind="pie", autopct="%1.1f%%")

plt.title("Pass vs Fail Distribution")

plt.ylabel("")
plt.savefig("C:/data-analysis-portfolio/visualizations/Pass vs Fail Distribution.png")

plt.show()
```

```python
gender_scores = df.groupby("sex")[['G1','G2','G3']].mean()

print(gender_scores)
```

```python
gender_scores.plot(kind="bar")

plt.title("Average Grades by Gender")

plt.ylabel("Score")
plt.savefig("C:/data-analysis-portfolio/visualizations/Average Grades by Gender.png")
plt.show()
```

```python
study_scores = df.groupby("studytime")[['G1','G2','G3']].mean()

print(study_scores)
```

```python
study_scores.plot(kind="bar")

plt.title("Impact of Study Time on Grades")

plt.ylabel("Average Grade")
plt.savefig("C:/data-analysis-portfolio/visualizations/Impact of Study Time on Grades.png")

plt.show()
```

```python
correlation = df[['G1','G2','G3','studytime','absences']].corr()

sns.heatmap(correlation, annot=True, cmap="coolwarm")

plt.title("Correlation Analysis")
plt.savefig("C:/data-analysis-portfolio/visualizations/Correlation Analysis.png")

plt.show()
```
